```python
import cv2

# open video file
cap = cv2.VideoCapture('video.mp4')

# get frame rate
fps = cap.get(cv2.CAP_PROP_FPS)

# set counter for file naming
count = 0

# loop over frames
while cap.isOpened():
    # read frame
    ret, frame = cap.read()

    # if there are no more frames, break out of loop
    if not ret:
        break

    # save frame as image
    cv2.imwrite(f'frame{count}.jpg', frame)

    # increment counter
    count += 1

    # wait for next frame
    cv2.waitKey(int(1000/fps))

# release video capture
cap.release()

# close all windows
cv2.destroyAllWindows()
```
